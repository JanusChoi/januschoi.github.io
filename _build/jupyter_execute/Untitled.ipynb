```python
import csv
import json
from datetime import datetime

def csv_to_sql_insert(csv_file_path):
    sql_inserts = []

    with open(csv_file_path, 'r', newline='', encoding='utf-8') as csvfile:
        reader = csv.DictReader(csvfile)
        for index, row in enumerate(reader):
            if index == 2:  # 读取5行后停止
                break
            # 解析tags字段
            tags = json.loads(row['tags'].replace("'", '"'))
            tag_name = ', '.join(tags)

            # 生成SQL插入语句
            sql_insert = f"""
INSERT INTO public.web_navigation (
    name, 
    title, 
    content, 
    detail, 
    url, 
    image_url, 
    thumbnail_url, 
    collection_time, 
    tag_name, 
    website_data, 
    star_rating, 
    category_name
) VALUES (
    '{row['name'].lower().replace(' ', '-')}',
    '{row['name']}',
    '{row['description']}',
    E'### What is {row['name']}?\\n\\n{row['description']}\\n\\n### How can I use {row['name']}?\\n\\nPlease visit {row['website']} for more information.',
    '{row['website']}',
    NULL,
    NULL,
    '{datetime.now().strftime('%Y-%m-%d %H:%M:%S')}',
    '{tag_name}',
    NULL,
    0,
    '{tags[0].lower() if tags else 'uncategorized'}'
);
"""
            sql_inserts.append(sql_insert)

    return sql_inserts
```

```python
# 使用示例
csv_file_path = '/Users/januswing/data/site4update.csv'  # 替换为您的CSV文件路径
sql_statements = csv_to_sql_insert(csv_file_path)

# 打印生成的SQL语句
for statement in sql_statements:
    print(statement)
```
